```python
import csv

def analyze_data(file_path):
    with open(file_path, newline='') as csvfile:
        csv_reader = csv.reader(csvfile, delimiter=',')
        headers = next(csv_reader)
        num_respondents = 0
        complete_respondents = 0

        question_groups = {}
        
        for header in headers:
            question_base = header.split('_')[0]
            if question_base not in question_groups:
                question_groups[question_base] = []
            question_groups[question_base].append(header)
        print(len(question_groups))
        for row in csv_reader:
            num_respondents += 1
            complete = True
            for group in question_groups.values():
                if not any(row[headers.index(part)] for part in group):
                    complete = False
                    break
            if complete:
                complete_respondents += 1

    return num_respondents, complete_respondents  

absfile_path = 'C:/Users/LENOVO/PycharmProjects/AI/LAB2/surveyDataSience.csv'
num_respondents, complete_respondents = analyze_data(absfile_path)

print(f'Numărul de respondenți: {num_respondents}')
print(f'Numărul de respondenți cu date complete: {complete_respondents}')
```
