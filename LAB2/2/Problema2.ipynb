```python
import os

image_folder = 'images'

for filename in os.listdir(image_folder):
    if filename.endswith(('.png', '.jpg', '.jpeg', '.webp')):
        image_path = os.path.join(image_folder, filename)
    with Image.open(image_path) as img:
        width, height = img.size
        print(f'Image: {filename}, Wight: {width}, Height= {height}')
```

```python
for filename in os.listdir(image_folder):
    if filename.endswith(('.png', '.jpg', '.webp')):
        image_path = os.path.join(image_folder, filename)
    with Image.open(image_path) as img:
        new_image = img.resize((128, 128))
        new_image.save(image_path)
        width, height = img.size
        print(f'Image: {filename}, Wight: {width}, Height= {height}')
```

```python
import matplotlib.pyplot as plt

image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.webp', '.png'))]

num_collumns = 3
num_rows = (len(image_files) + num_collumns - 1) // num_collumns

fig, axes = plt.subplots(num_rows, num_collumns, figsize=(15, 5 * num_rows))

axes = axes.flatten()

for ax, image_file in zip(axes, image_files):
    image_path = os.path.join(image_folder, image_file)
    img = Image.open(image_path).convert('RGB')
    ax.imshow(img)
    ax.set_title(image_file)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
from PIL import ImageFilter

image_pathc = 'images/LeCun.jpg'
original_image = Image.open(image_pathc).convert('RGB')

blurred_image = original_image.filter(ImageFilter.GaussianBlur(radius=5))

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

axes[0].imshow(original_image)
axes[0].set_title("Before")
axes[0].axis('off')

axes[1].imshow(blurred_image)
axes[1].set_title('After')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image_path = 'images/Russell.jpg'

image = cv2.imread(image_path, cv2.IMREAD_COLOR_RGB)

edges = cv2.Canny(image, threshold1=100, threshold2=300)

fig1, axes = plt.subplots(1, 2, figsize=(15, 6))

axes[0].imshow(image)
axes[0].set_title('Before')
axes[0].axis('off')

axes[1].imshow(edges, cmap='gray')
axes[1].set_title('After')
axes[1].axis('off')

plt.tight_layout()
plt.show()


```
