```python
from icrawler.builtin import BingImageCrawler

def download_images(keyword, num_images, folder):
    crawler = BingImageCrawler(storage={"root_dir": folder})
    crawler.crawl(keyword=keyword, max_num=num_images)

download_images("bike", 300, "./images/bikes")

```

```python
import os

def rename_images(folder_path):

    files = sorted(f for f in os.listdir(folder_path) if f.startswith("000") and f.lower().endswith(('.jpg', '.png', '.jpeg')))

    for index, filename in enumerate(files, start=50):
        ext = os.path.splitext(filename)[1] 
        new_name = f"car{index}{ext}"
        old_path = os.path.join(folder_path, filename)
        new_path = os.path.join(folder_path, new_name)
        
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} -> {new_name}")

rename_images("images/cars")
```
